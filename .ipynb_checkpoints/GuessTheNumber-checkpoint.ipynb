```python
import random
 
# print("Welcome to:")
print(" ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____ ")       
print("(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)")      
print("  ___  _  _  ____  ____  ____        ____  _  _  ____        __ _  _  _  _  _  ____  ____  ____ ")       
print(" / __)/ )( \\(  __)/ ___)/ ___)      (_  _)/ )( \\(  __)      (  ( \\/ )( \\( \\/ )(  _ \\(  __)(  _ \\")       
print("( (_ \\) \\/ ( ) _) \\___  \\___ \\        )(  ) __ ( ) _)       /    /) \\/ (/ \\/ \\ ) _ ( ) _)  )   /")      
print(" \\___/\\____/(____)(____/(____/       (__) \\_)(_/(____)      \\_)__)\\____/\\_)(_/(____/(____)(__\\_)")       
print(" ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____  ____ ")       
print("(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)(____)")
 
print("\nLoading game...\n ")
name = input("\nHello! What is your name? ")
 
seed_random = int(input("\nWhat is your favorite number: "))
# Set the seed for random
random.seed(seed_random)
 
MAX_TRY = 7
MAX_NUM = 100
MIN_NUM = 1
 
#Does not account for other inputs
 
yesNoCheck = False
#Temporary solution, not sure what Zybook wants
while yesNoCheck == False:
    yes_no = input('\nDo you want to check instructions? [y/n] or [Y/N]')
    if yes_no == 'y' or yes_no == 'Y':
        print('\n\nInstructions on how to play:')
        print('I am thinking of a whole number from 1 to 100')
        print('Try to guess my number. After your guess, I will show two tips')
        print('Tip 1: If your guess is low, high or invalid compared to my number')
        print('Tip 2: I will type one or more stars (*). The more stars I type, \nthe closer you are to my number.')
        print('(1 star (*) means far away, 7 stars (*******) means really close)\nYou get 7 guesses.')
        print()
        print()
        print(f"Let's play {name} ")
        yesNoCheck = True
    elif yes_no == 'n' or yes_no == 'N':
        print(f"\nLet's play {name} ")
        yesNoCheck = True
 
print(f"\nComputer is thinking...")
print(f"\n{name}, you have 7 tries to guess a number between 1 and 100.\n")
 
num = random.randrange(MIN_NUM, MAX_NUM)
print(num)
attempt = 1
check = False
while (check == False) and (attempt <= MAX_TRY):
    print(f"Try {attempt}")
    guess = int(input("Enter a guess: "))
    delta = abs(guess - num)
    if (delta == 1):
        star = '*******'
    elif (delta >= 2) & (delta <= 3):
        star = '******'
    elif (delta >= 4) & (delta <= 7):
        star = '*****'
    elif (delta >= 8) & (delta <= 15):
        star = '****'
    elif (delta >= 16) & (delta <= 35):
        star = '***'
    elif (delta >= 36) & (delta <= 55):
        star = '**'
    else :
        star = '*'
    if (guess < MIN_NUM) | (guess > MAX_NUM):
        print("You must enter a number between 1 and 100.")
    elif guess == num:
        print(f"\nGood Job, {name}! You guessed my number in {attempt} guess{'es' if attempt > 1 else ''}!")
        check = True
    else:
        attempt += 1
        if guess > num:
            print("Guess too high!")
            print(f"{star}\n")
        else:
            print("Guess too low!")
            print(f"{star}\n")
if check == False:
    print("Sorry, you have reached the maximum tries.\nGame over...")
```
