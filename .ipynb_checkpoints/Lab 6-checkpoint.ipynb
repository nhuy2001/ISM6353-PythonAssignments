**Question 7**
<br>
Create a dataframe that shows the number of records for each category in 2016 and 2017 using order date.
<br>
i.e. Each row should show the number of records for a category in 2016 and 2017. 
<br>
There should be 3 columns: Category, 2016_Records, and 2017_Records. Sort the dataframe by category. 
<br>
What’s the value at index [10, 2]?

```python
Q7dat = superstore.copy()
Q7dat['Year'] = pd.to_datetime(Q7dat['Order Date'], format = '%m/%d/%Y').dt.year

CatYear = Q7dat\
    [(Q7dat['Year'] == 2016) | (Q7dat['Year'] == 2017)].\
    groupby(['Year', 'Category']).size().to_frame().\
    rename(columns = {0: 'Count'}).\
    reset_index()

print(f'CatYear: \n{CatYear.head(10)}')
```

```python
#Pivot Wider
CatYear_Pivot = CatYear.\
    pivot(index = 'Category', columns = 'Year', values = 'Count').\
    reset_index().\
    sort_values('Category').\
    rename(columns = {2016: '2016_Records', 2017: '2017_Records'}).\
    rename_axis(None, axis = 1)

print(f'CatYear_Pivot @ [10,2]: {CatYear_Pivot.iloc[10,2]}\n')
print(f'CatYear_Pivot: \n{CatYear_Pivot.head(10)}')
```

**Question 15**
<br>
Use breast cancer data. Run an ICA for 1 to 30 components.
<br>
Use max_iter = 1000, random_state = 101. Create a figure of kurtosis as the number of components increases.
<br>
Upload the figure. 

```python
# extrac x and y vars 
components = list(range(1, features.shape[1]+1))
kurtosis = []

for i in range(1, xQ11.shape[1]+1): 
    
    # define ICA
    ica = FastICA(n_components = i, max_iter = 1000, random_state = 101)
    
    # fit ICA to data 
    ica.fit(xQ11)
    
    # transform x-vars using ind components
    projected_variables = ica.transform(xQ11)
    projected_variables = pd.DataFrame(projected_variables)
    
    # calc. mean absolute value of kurtosis for the components
    k = projected_variables.kurt(axis = 0).abs().mean()
    kurtosis.append(k)
    
    # using kurtosis is good if your goal is to find # of independent components rather than making predictions
    print('components:', i, ', kurtosis:', k)
    
plt.plot(components, kurtosis)
plt.ylabel('Kurtosis')
plt.xlabel('# of Components')
plt.title('Kurtosis vs. # of Independent Components')
plt.savefig('Q15_Plot (Lab 6).png', dpi = 600)
```
