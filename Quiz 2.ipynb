3. Use the Superstore data. What's the item ID with the most quantity ordered in year 2016? 

```python
superstore['Year'] = pd.to_datetime(superstore['Order Date'], format = '%m/%d/%Y').dt.year
```

```python
superstore[superstore.Year == 2016].groupby('Item ID').sum(numeric_only = True)['Order Quantity'].sort_values(ascending = False).head(5)
```

Merge this dataframe with the original dataframe using state as the key, so a new column Total_State_Sales showing total sales for the state in each row is added.

What's the value at index [500, 27]? Round to nearest 1 decimal.
