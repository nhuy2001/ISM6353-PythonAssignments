### Lambda Functions

```python
def tip_conversion(x): 
    if x.tip < 3: 
        tip_level = 'small'
    elif x.tip < 5:
        tip_level = 'medium'
    else: 
        tip_level = 'high'
        
    return tip_level 

data['tip_level'] = data.apply(lambda x: tip_conversion(x), axis = 1)

data
```
