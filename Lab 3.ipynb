```python
# 9. Use sklearn. Train the following logistic regressions with x-variables:

# Model 1: Age, Daily Time Spent on Site
# Model 2: Age, Daily Time Spent on Site, Area Income, Male
# Model 3: Age, Daily Time Spent on Site, Area Income, Male, Daily Internet Usage

# Use test_size=0.3, random_state=101 for all models.

# Which model has the highest AUC for test data? Enter the model number.
```

```python
#Custom Function
def classification_metrics(Y_pred, Y_true):
    acc = accuracy_score(Y_true, Y_pred)
    precision = precision_score(Y_true, Y_pred)
    recall = recall_score(Y_true, Y_pred)
    f1score = f1_score(Y_true, Y_pred)
    auc = roc_auc_score(Y_true, Y_pred)

    return acc, precision, recall, f1score, auc

def display_metrics(classifierName, Y_pred, Y_true):
    print ("______________________________________________")
    print ("Model: "+classifierName)
    acc, precision, recall, f1score, auc = classification_metrics(Y_pred, Y_true)
    # returns 5 vars: acc, precision, recall, f1score, auc
    # print them below
    print ("Accuracy: "+str(acc))
    print ("Precision: "+str(precision))
    print ("Recall: "+str(recall))
    print ("F1-score: "+str(f1score))
    print ("AUC: "+str(auc))
    print ("______________________________________________")
    print ("")
```

```python
mod1x = advertising[['Age', 'Daily Time Spent on Site']]
mod2x = advertising[['Age', 'Daily Time Spent on Site', 'Area Income', 'Male']]
mod3x = advertising[['Age', 'Daily Time Spent on Site', 'Area Income', 'Male', 'Daily Internet Usage']]

y = advertising['Clicked on Ad']
```

```python
x1Q9tr, x1Q9te, y1Q9tr, y1Q9te = train_test_split(mod1x, y, test_size=0.3, random_state=101)
x2Q9tr, x2Q9te, y2Q9tr, y2Q9te = train_test_split(mod2x, y, test_size=0.3, random_state=101)
x3Q9tr, x3Q9te, y3Q9tr, y3Q9te = train_test_split(mod3x, y, test_size=0.3, random_state=101)
```

```python
#Since I reused matrix3 for Q11, it messes up the results for Q9 since the other two models (i.e., mod1 and mod2) don't use liblinear. Consequently, to keep thing simple, I used liblinear for mod1 and mod2
mod1 = LogisticRegression(solver = 'liblinear')
mod1.fit(x1Q9tr, y1Q9tr)

mod2 = LogisticRegression(solver = 'liblinear')
mod2.fit(x2Q9tr, y2Q9tr)

mod3 = LogisticRegression(solver = 'liblinear')
mod3.fit(x3Q9tr, y3Q9tr)
```

```python
y1Pred = mod1.predict(x1Q9te)
y2Pred = mod2.predict(x2Q9te)
y3Pred = mod3.predict(x3Q9te)

print('y1Pred: \n', y1Pred, '\n',
      'y2Pred: \n', y2Pred, '\n',
      'y3Pred: \n', y3Pred)
```

```python
matrix1 = confusion_matrix(y1Q9te, y1Pred)
print('Confusion Matrix: \n', matrix1)
display_metrics('Logistic Regression Model 1', y1Pred, y1Q9te)
```

```python
matrix2 = confusion_matrix(y2Q9te, y2Pred)
print('Confusion Matrix: \n', matrix2)
display_metrics('Logistic Regression Model 2', y2Pred, y2Q9te)
```

```python
matrix3 = confusion_matrix(y3Q9te, y3Pred)
print('Confusion Matrix: \n', matrix3)
display_metrics('Logistic Regression Model 3', y3Pred, y3Q9te)
```

```python
# 12. Use sklearn. Train 5 logistic regressions with x-variables: Age, Daily Time Spent on Site, Area Income, Male, Daily Internet Usage
# Vary the test size for each model (use 30%, 40%, 50%, 60%, 70%). Use random_state=101.

# What is the average accuracy for the 5 models for the test set? Round to 2 decimals.
```

```python
#Preparation
accuracies = []
#testSizes = np.arange(0.3, 0.8, 0.1) #array
testSizes = [0.3, 0.4, 0.5, 0.6, 0.7]
testSizes
```

```python
#For loop
modQ12 = LogisticRegression(solver = 'liblinear')
for i in testSizes:
    xQ12tr, xQ12te, yQ12tr, yQ12te = train_test_split(mod3x, y, test_size = i, random_state = 101)
    modQ12.fit(xQ12tr, yQ12tr)
    yQ12Pred = modQ12.predict(xQ12te)
    
    acc, precision, recall, f1score, auc = classification_metrics(yQ12Pred, yQ12te)
    accuracies.append(acc)
```
