```python
##Data Preparation
#Perform all calculations beforehand
df['unit_profit'] = df.price - df.cost
df['sales'] = df.price * df.quantity
df['total_cost'] = df.price * df.cost
df['total_profit'] = df.sales - df.total_cost

#Convert date column to datetime datatype
df['date'] = df.date.astype('datetime64[ns]')

df
```

```python
#15. Create a new variable 'total_cost', which is price mulitpled by cost
#Create a figure that contains 2 line plots (NOT as 2 separate subplots).
#The 2 line plots are: Sales vs. Date (blue, dashdot) and Total Cost vs. Date (green, dashed)
#X-axis can be labeled as Date, and Dollars ($) should be on the y-axis
```

```python
q15plot = (
    p9.ggplot(df, mapping = p9.aes(x = 'date'))+
    #Sales
    p9.geom_line(mapping = p9.aes(y = 'sales', color = "'blue'"), linetype = '-.')+
    #Total Cost
    p9.geom_line(mapping = p9.aes(y = 'total_cost', color = "'green'"), linetype = '--')+
    
    #Other Settings
    p9.theme_classic()+
    p9.theme(axis_text_x = p9.element_text(rotation = 90, hjust = 1))+
    p9.labs(title = 'Sales and Total Cost over Two Years')+
    p9.scale_color_identity(guide='legend',name='Metric',
                        breaks=['blue','green'],
                        labels=['sales','total_cost'])+
    p9.scale_x_datetime(name = 'Date')+
    p9.scale_y_continuous(name = 'Dollars ($)')
             )
q15plot
```

```python
q15plot.save(filename = 'fig2.png', dpi=600)
```
